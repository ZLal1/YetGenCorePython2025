# Problem 1

Şimdi beraber kullanıcılardan aldığımız verileri işleyerek, onlara hem çıktı verecek, hem de aldığımız ve hesapladığımız verileri ileride kullanabilmek adına kaydedeceğiz.


 **Bu bir veri giriş platformu olacak**


 * kullanıcıya kaç giriş yapacağını soracağız
* kullanıcıdan isim-soyisim, yaş, boy, kilo bilgilerini alarak VKİ hesaplayacağız
* ardından bunları sınıflara göre kaydedeceğiz bir dosyaya kaydedeceğiz


Sınıflar ise şu şekilde: 
* 18, 5 kg/m.'nin altında olanlar: Zayıf
* 18.5 – 24, 9 kg/m. arasında olanlar: Normal kilolu
* 25 – 29, 9 kg/m. arasında olanlar: Fazla kilolu
* 30 – 39, 9 kg/m. arasında olanlar: Obez
* 40 kg/m.'nin üzerinde olanlar: İleri derecede obez (morbid obez), olarak görülür.

```python
# 1.Problemin Çözümünü Buraya Yazınız

def vki_hesapla(kilo,boy):
    return (kilo / (boy ** 2))
            
def vki_sınıfları(vki):
    if vki < 18.5:
        return "Zayıf"
    elif 18.5 <= vki <= 24.9:
        return "Normal Kilolu"
    elif 25 <= vki <= 29.9:
        return "Fazla Kilolu"
    elif 30 <= vki <= 39.9:
        return "Obez"
    else:
        return "İleri Derecede Obez (Morbid Obez)"
    
def veri_giris_platformu ():
    kayitlar = []
    kisi_sayisi = int(input("Kaç giriş yapacaksınız?"))

    for i in range (kisi_sayisi):
        print(f"\n{i+1}. kişi için bilgileri giriniz: ")
        isim = input("İsim:")
        soyisim = input("Soyisim:")
        yas = int(input("Yaş: "))
        boy = float(input("Boy(m): "))
        kilo = float(input("Kilo(kg): "))

        vki = vki_hesapla(kilo,boy)
        sinif = vki_sınıfları(vki)
        print(f"Vücut Kitle İndeksiniz: {vki:.2f} ({sinif})")

        kayit = {
            "Isim": isim,
            "Soyisim": soyisim,
            "Yaş": yas,
            "Boy(m)": boy,
            "Kilo(kg)": kilo,
            "VKİ": round (vki , 2),
            "Sınıf": sinif
        }
        kayitlar.append(kayit)

    dosya_adi = "vki_kayitlari.txt"

    with open("vki_kayitlari.txt", "r+", encoding = "utf-8") as file:
        file.write("Isim Soyisim | Yaş | Boy(m) | Kilo(kg) | VKİ | Sınıf\n")
        file.write("--------------------" + "\n")
        for kayit in kayitlar:
            satir = f"{kayit['Isim']} {kayit['Soyisim']} | {kayit['Yaş']} | {kayit['Boy(m)']:.2f} | {kayit['Kilo(kg)']:.1f} | {kayit['VKİ']:.2f} | {kayit['Sınıf']}\n"
            file.write(satir)

    print(f"\nTüm veriler '{dosya_adi}' dosyasına kaydedildi.")

if __name__ == "__main__":
    veri_giris_platformu()
```

# Problem 2
**YetBar - Smoothie Bank**
---
YetBar spor salonunun girişinde bulunan, insansız bir sağlıklı içecek otomatıdır. Kullanıcı çeşitli girdiler smoothie içeceğine koyacağı ürünleri belirleyebilir, fiyat hesaplaması yaptırabilir ya da sadece menüyü görüntüleyebilir.

**Projenin Amacı:**

*   Birbirini etkileyen ve bir kod içerisinde birden fazla defa başvurulacak fonksiyonların oluşturulup, "fonksiyonel" proglamlamak.
*   Büyük programları yazarken, yazılımcı olarak yapabileceğimiz hatalardan kaçınmak için kontrol yöntemleri geliştirmek.
*   Kullanıcı inputlarını ve yaptığımız hesapları, raporlamak/loglamak amacıyla kaydetmek.

**Programdan Beklenen Çıktılar**
* Program kullanıcıya temelde 2 farklı işlem yapma imkanı sağlar.
> 1. Menüyü görüntüleme
> 3. Satın alma
* Satın alınan her içecek tarih, fiyat bilgisi ile kaydedilmeli ve her zaman ulaşılabilir olmalıdır.




```python
# 2.Problemin Çözümünü Buraya Yazınız
import time

menu = {
    "elma": 0.50,
    "uzum": 0.75,
    "armut": 0.90,
    "çilek": 0.95,
    "muz": 1,
    "yabanmerisini": 1.50,
    "ananas": 2
}

print("Elma Fiyatı:", menu["elma"])
print("Ananas Fiyatı:", menu["ananas"])

def display_menu():
    print("---------YetBar MENU---------")
    for item in list(menu.keys()):
        print(f"{item.upper():<10}: {menu[item]} TL")

def tek_urun_fiyat_hesaplama(urun, adet):
    assert adet >= 0, "Ürün adedi 0 veya 0'dan küçük olamaz."
    if urun not in list(menu.keys()):
        return None
    else:
        return menu[urun] * adet

def satin_alma():
    display_menu()
    print("Lütfen satın almak istediğiniz ürünü giriniz. Çıkış yapmak için 0 tuşuna basınız.")

    satın_alınan_ürünler = []
    satın_alma_adedi = []
    son_fiyat = 0

    while True:
        eklenecek_urun = input("Hangi üründen eklemek istiyorsunuz? ")
        if eklenecek_urun == "0":
            print("Fiş kaydediliyor...")
            break
        ekleme_adedi = int(input("Kaç adet eklemek istiyorsunuz? "))
        fiyat = tek_urun_fiyat_hesaplama(eklenecek_urun, ekleme_adedi)
        if fiyat is None:
            print("Yanlış girdiniz. Ürün ekleme başarısız.")
        else:
            print(f"Satın alma ücreti: {fiyat} TL.\nSatın almayı onaylamak için 'y', iptal için 'x' giriniz.")
            check = input("Satın almayı onaylıyor musunuz? ")
            if check == "y":
                print("Satın alım onaylandı.")
                print("--------------------")
                son_fiyat += fiyat
                satın_alınan_ürünler.append(eklenecek_urun)
                satın_alma_adedi.append(ekleme_adedi)
                print(f"Ara Toplam: {son_fiyat} TL")
            elif check == "x":
                print("Satın alma işlemi iptal edildi.")
                print("----------------")
    return son_fiyat, satın_alınan_ürünler, satın_alma_adedi

def main():
    print("YetBar'a hoş geldiniz!\nMenüyü görüntülemek için '1' e, smoothie yapmak için '2' ye basabilirsiniz. Çıkmak için '0' a basınız.")
    program = None
    while True:
        program = int(input("Tercihiniz: "))
        if program < 0 or program > 2:
            print("Yanlış bir tercih yaptınız. Lütfen yeniden deneyiniz.")
        elif program == 0:
            print("Çıkış yapılıyor...\nMutlu günler!")
            break
        elif program == 1:
            display_menu()
        elif program == 2:
            satin_alma_ücreti, satin_alinan_urunler, satin_alma_adedi = satin_alma()
            with open("kasa.txt", "a", encoding="utf-8") as file:
                file.write(str(time.ctime(time.time())))
                file.write("\n")
                file.write(','.join(map(str, satin_alinan_urunler)))
                file.write("\n")
                file.write(','.join(map(str, satin_alma_adedi)))
                file.write("\n")
                file.write(f"Toplam: {satin_alma_ücreti} TL")
                file.write("\n")
                file.write("-----------------")
                file.write("\n")
            cikis_kontrol = input("Başka bir işlem yapmak ister misiniz? (evet/hayır) ")
            if cikis_kontrol.lower() == "evet":
                pass
            else:
                print("\n\nTeşekkürler.")
                break

if __name__ == "__main__":
    main()






```
