# Problem 1

Kumanda isminde bir sınıf oluşturarak aşağıdaki işlemleri yapmasını sağlayan fonksiyonları sınıf içerisinde yazalım.
1. Tv Aç
2. Tv Kapat
3. Ses Ayarları
4. Kanal Ekle
5. Kanal Sayısını Öğrenme
6. Rastgele Kanala Geçme
7. Televizyon Bilgileri


Kumandayı kapatmak için q tuşuna basılmalı ve diğer durumlarda hangi işlemi yapmak istediği sorulmalıdır. Her işlemde gerekli fonksiyonunun sınıf içerisinden çağırılması gerekmektedir.

random ve time kütüphanesini araştırarak kullanabilirsiniz.

Kendi istediğiniz özellikleri eklemekte serbetsiniz.

```python
# 1.Problemin Çözümünü Buraya Yazınız

import random
import time


class Kumanda():
    def __init__(self):
        self.acik_mi = False
        self.ses = 5
        self.kanallar = ["NOW TV", "TV 8", "Show TV", "TLC", "DMAX"]
        self.suanki_kanal = "NOW TV"
        self.bilgiler = {
            "Marka": "LG",
            "Çıkış Yılı": 2024,
            "Çözünürlük": "3840 x 2160 piksel"
        }
    def tv_ac(self):
        if self.acik_mi:
            print("TV zaten açık.")

        else:
            self.acik_mi = True
            print("TV açılıyor...")

    def tv_kapat(self):
        if self.acik_mi:
            print("TV zaten kapalı.")
        else:
            print("TV kapatılıyor...")

    def ses_ayarlari(self):
        while True:
            secim = input("Sesi azalt: '<'\nSesi artır: '>'\nÇıkmak için: 'q'\nSeçiminiz: ")
            if secim == "<":
                if self.ses > 0:
                    self.ses -= 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten en düşük seviyede.")
            elif secim == ">":
                if self.ses < 100:
                    self.ses += 1
                    print(f"Ses: {self.ses}")
                else:
                    print("Ses zaten en yüksek seviyede.")
            elif secim == "q":
                break
            else:
                print("Geçersiz tuşlama.")

    def kanal_ekle(self):
        yeni_kanal = input("Yeni kanal adını girin: ")
        self.kanallar.append(yeni_kanal)
        print(f"{yeni_kanal} kanalı eklendi.")

    def kanal_sayisi(self):
        print(f"Toplam kanal sayısı: {len(self.kanallar)}")

    def rastgele_kanal(self):
        yeni_kanal = random.choice(self.kanallar)
        self.aktif_kanal = yeni_kanal
        print(f"Rastgele kanala geçildi: {self.aktif_kanal}")

    def tv_bilgileri_goster(self):
        print("Televizyon Bilgileri:")
        for key, value in self.bilgiler.items():
            print(f"{key}: {value}")

    def kumanda_kapat(self):
        print("Kumanda kapanıyor...")
        time.sleep(1)
        print("Kumanda kapalı.")

    def menu(self):
        while True:
            print("\n--- Kumanda Menü ---")
            print("1. TV Aç")
            print("2. TV Kapat")
            print("3. Ses Ayarları")
            print("4. Kanal Ekle")
            print("5. Kanal Sayısını Öğrenme")
            print("6. Rastgele Kanala Geçme")
            print("7. Televizyon Bilgileri")
            print("q. Kumandayı Kapat")

            secim = input("Yapmak istediğiniz işlemi seçin (1-7 veya q): ")

            if secim == "1":
                self.tv_ac()
            elif secim == "2":
                self.tv_kapat()
            elif secim == "3":
                self.ses_ayarlari()
            elif secim == "4":
                self.kanal_ekle()
            elif secim == "5":
                self.kanal_sayisi()
            elif secim == "6":
                self.rastgele_kanal()
            elif secim == "7":
                self.tv_bilgileri_goster()
            elif secim == "q":
                self.kumanda_kapat()
                break
            else:
                print("Geçersiz seçim, tekrar deneyin.")

# Ana program
kumanda = Kumanda()
kumanda.menu()



```
